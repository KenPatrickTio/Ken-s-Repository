````python
## Code goes here
import pandas as pd
import csv
import re
````

````python
df = pd.read_csv("recweekdata.csv")
````

````python
with open("studentprofiles.csv",newline="") as csv_file:
    studentprofiles = csv.DictReader(csv_file)
    gender_list = []
    for i in studentprofiles:
        gender_list.append(i["gender"])
    df["gender"] = gender_list
````

````python
# Lamda Replace for BirthMonth
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Dec$","December", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Nov$","November", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Oct$","October", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Sep$","September", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Aug$","August", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Jul$","July", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Jun$","June", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("May$","May", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Apr$","April", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Mar$","March", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Feb$","February", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub("Jan$","January", x) )

df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^12","December", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^11","November", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^10","October", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^9","September", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^8","August", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^7","July", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^6","June", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^5","May", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^4","April", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^3","March", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^2","February", x) )
df.loc[:,"birthmonth"] = df.loc[:,"birthmonth"].apply(lambda x: re.sub(r"^1","January", x) )
````

````python
# Lamda Replace for BirthYear
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].astype(str).apply(lambda x: re.sub("1997","97", x) )
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].astype(str).apply(lambda x: re.sub("1998","98", x) )
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].astype(str).apply(lambda x: re.sub("1999","99", x) )
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].astype(str).apply(lambda x: re.sub("2000","00", x) )
df.loc[:,"birthyear"] = df.loc[:,"birthyear"].astype(str).apply(lambda x: re.sub("2001","01", x) )
````

````python
# Lamda Replace for Course
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^BS LM","BS Legal Management", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^Management","BS Management", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^BS CTM","BS Comtech", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^BS ITE","BS IT Entrepreneurship", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^Management Engineering","BS Management Engineering", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^CTM","BS Comtech", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^BS ME","BS Management Engineering", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^BS MGT","BS Management", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^Information Technology Entrepreneurship","BS IT Entrepreneurship", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^LM","BS Legal Management", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^ITE","BS IT Entrepreneurship", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^MGT","BS Management", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^ME","BS Management Engineering", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^Communications Technology","BS Comtech", x) )
df.loc[:,"course"] = df.loc[:,"course"].apply(lambda x: re.sub("^Legal Management","BS Legal Management", x) )
````

````python
# Lamda Replace for Phonenumber
df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].apply(lambda x: re.sub("^917","+63 917", x) )
df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].apply(lambda x: re.sub("^918","+63 918", x) )
df.loc[:,"phonenumber"] = df.loc[:,"phonenumber"].apply(lambda x: re.sub("^927","+63 927", x) )

# Sir's Lamda apply for Carrier
carrier = {"+63 917":"Globe","+63 918":"Smart", "+63 927":"Globe"}
df["carrier"] = df.loc[:,"phonenumber"].str[0:7].apply(lambda x: carrier[x])

df
````
